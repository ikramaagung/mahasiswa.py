```python
class Mobil:
    def __init__(self, merk, model):
        self.merk = merk
        self.model = model

    def nyalakan_mesin(self):
        print(f"Mobil {self.merk} {self.model} sedang dinyalakan...")

# Membuat objek
avanza = Mobil("Toyota", "Avanza")

# Memanggil metode
avanza.nyalakan_mesin()
```
